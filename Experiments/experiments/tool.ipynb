```python
import json
import numpy as np
import matplotlib.pyplot as plt

def calculate_average(json_file):
    with open(json_file, 'r') as f:
        data = json.load(f)
    
    post_positive = []
    post_negative = []
    post_neutral = []
    transferred_post_positive = []
    transferred_post_negative = []
    transferred_post_neutral = []

    for item in data:
        post_positive.append(item['Post_Postive'])
        post_negative.append(item['Post_Negative'])
        post_neutral.append(item['Post_Neutral'])
        transferred_post_positive.append(item['Transferred_Post_Postive'])
        transferred_post_negative.append(item['Transferred_Post_Negative'])
        transferred_post_neutral.append(item['Transferred_Post_Neutral'])

    post_positive_avg = np.mean(post_positive)
    post_negative_avg = np.mean(post_negative)
    post_neutral_avg = np.mean(post_neutral)
    transferred_post_positive_avg = np.mean(transferred_post_positive)
    transferred_post_negative_avg = np.mean(transferred_post_negative)
    transferred_post_neutral_avg = np.mean(transferred_post_neutral)

    return (post_positive_avg, post_negative_avg, post_neutral_avg, transferred_post_positive_avg, transferred_post_negative_avg, transferred_post_neutral_avg)

def plot_bar_chart(averages):
    labels = ['P_Pos', 'P_Neg', 'P_Neu', 'TP_Pos', 'TP_Neg', 'TP_Neu']
    values = averages

    fig, ax = plt.subplots()
    ax.bar(labels[:3], values[:3], color='blue', label='Post')
    ax.bar(labels[3:], values[3:], color='orange', label='Transferred_Post')
    ax.set_ylabel('Average Score')
    ax.set_title('Average Sentiment Scores')
    ax.legend()
    plt.show()

json_file = "/home/qiang/projects/Digital_mental_health/final_dataset/experiments/datasets/SAS2UAS_sentiment.json"
averages = calculate_average(json_file)
print("Average Scores:")
print("Post Positive:", averages[0])
print("Post Negative:", averages[1])
print("Post Neutral:", averages[2])
print("Transferred Post Positive:", averages[3])
print("Transferred Post Negative:", averages[4])
print("Transferred Post Neutral:", averages[5])

plot_bar_chart(averages)
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np

def process_json_file(input_file):
    with open(input_file, 'r') as f:
        data = json.load(f)

    post_counts = {'Positive': 0, 'Negative': 0, 'Neutral': 0}
    transferred_post_counts = {'Positive': 0, 'Negative': 0, 'Neutral': 0}

    for item in data:
        post_positive = item['Post_Postive']
        post_negative = item['Post_Negative']
        post_neutral = item['Post_Neutral']
        transferred_post_positive = item['Transferred_Post_Postive']
        transferred_post_negative = item['Transferred_Post_Negative']
        transferred_post_neutral = item['Transferred_Post_Neutral']

        max_post_label = max(post_positive, post_negative, post_neutral)
        max_transferred_post_label = max(transferred_post_positive, transferred_post_negative, transferred_post_neutral)

        if max_post_label == post_positive:
            post_counts['Positive'] += 1
        elif max_post_label == post_negative:
            post_counts['Negative'] += 1
        else:
            post_counts['Neutral'] += 1

        if max_transferred_post_label == transferred_post_positive:
            transferred_post_counts['Positive'] += 1
        elif max_transferred_post_label == transferred_post_negative:
            transferred_post_counts['Negative'] += 1
        else:
            transferred_post_counts['Neutral'] += 1

    return post_counts, transferred_post_counts

def plot_bar_chart(post_counts, transferred_post_counts):
    labels = list(post_counts.keys())
    post_values = list(post_counts.values())
    transferred_post_values = list(transferred_post_counts.values())

    x = np.arange(len(labels))  # 
    width = 0.35  # 

    fig, ax = plt.subplots(figsize=(10, 6))  # 
    rects1 = ax.bar(x - width/2, post_values, width,color='blue',label='Post')
    rects2 = ax.bar(x + width/2, transferred_post_values, width, color='orange',label='Transferred_Post')

    ax.set_ylabel('Counts')
    ax.set_title('Counts of Sentiment Labels')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()

    def autolabel(rects):
        """Attach a text label above each bar in *rects*, displaying its height."""
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')

    autolabel(rects1)
    autolabel(rects2)

    fig.tight_layout()

    plt.show()

input_file = "/home/qiang/projects/Digital_mental_health/final_dataset/experiments/datasets/GAS2SPAS_sentiment.json"
post_counts, transferred_post_counts = process_json_file(input_file)
plot_bar_chart(post_counts, transferred_post_counts)
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np

def process_json_file(input_file):
    with open(input_file, 'r') as f:
        data = json.load(f)

    post_counts = {'Positive': [], 'Negative': [], 'Neutral': []}
    transferred_post_counts = {'Positive': [], 'Negative': [], 'Neutral': []}

    for item in data:
        post_positive = item['Post_Postive']
        post_negative = item['Post_Negative']
        post_neutral = item['Post_Neutral']
        transferred_post_positive = item['Transferred_Post_Postive']
        transferred_post_negative = item['Transferred_Post_Negative']
        transferred_post_neutral = item['Transferred_Post_Neutral']

        post_counts['Positive'].append(post_positive)
        post_counts['Negative'].append(post_negative)
        post_counts['Neutral'].append(post_neutral)
        transferred_post_counts['Positive'].append(transferred_post_positive)
        transferred_post_counts['Negative'].append(transferred_post_negative)
        transferred_post_counts['Neutral'].append(transferred_post_neutral)

    return post_counts, transferred_post_counts

def plot_bar_chart(post_counts, transferred_post_counts):
    labels = list(post_counts.keys())
    post_values = [np.mean(post_counts[label]) for label in labels]
    transferred_post_values = [np.mean(transferred_post_counts[label]) for label in labels]

    x = np.arange(len(labels))
    width = 0.35

    fig, ax = plt.subplots(figsize=(10, 5))

    rects1 = ax.bar(x - width/2, post_values, width, color='blue', label='Post')
    rects2 = ax.bar(x + width/2, transferred_post_values, width, color='orange', label='Transferred_Post')

    ax.set_ylabel('Counts')
    ax.set_title('Comparison of Sentiment Labels (Post vs Transferred_Post)')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()

    plt.tight_layout()
    plt.show()

input_file = "/home/qiang/projects/Digital_mental_health/final_dataset/experiments/datasets/SAS2UAS_sentiment.json"
post_counts, transferred_post_counts = process_json_file(input_file)
plot_bar_chart(post_counts, transferred_post_counts)
```
