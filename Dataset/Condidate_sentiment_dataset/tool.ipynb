```python
import csv

# 输入和输出文件的名称
input_filename = '/Users/qiangzhou/Library/CloudStorage/OneDrive-UniversityofPittsburgh/projects/Digital_mental_health/sentiment_dataset/tweet-sentiment-extraction/test.csv'
output_filename = 'negative_sentiments_test.csv'

# 打开输入文件并读取内容
with open(input_filename, mode='r', encoding='utf-8') as infile:
    reader = csv.reader(infile)
    # 获取列名
    headers = next(reader)
    
    # 打开输出文件准备写入
    with open(output_filename, mode='w', encoding='utf-8', newline='') as outfile:
        writer = csv.writer(outfile)
        # 写入列名
        writer.writerow(headers)
        
        # 遍历输入文件中的每一行
        for row in reader:
            # 检查第四列（假设列索引从0开始）的值是否为'negative'
            if row[2] == 'negative':
                # 如果是，写入到输出文件中
                writer.writerow(row)

print(f"Negative sentiment rows have been saved to '{output_filename}'.")
```

```python
import csv
import json

# 输入CSV文件和输出JSON文件的名称
input_csv_filename = '/Users/qiangzhou/Library/CloudStorage/OneDrive-UniversityofPittsburgh/projects/Digital_mental_health/sentiment_dataset/negative_sentiments_train.csv'
output_json_filename = 'negative_event_train.json'

# 初始化一个空列表来存储转换后的数据
data = []

# 打开并读取CSV文件
with open(input_csv_filename, mode='r', encoding='utf-8') as csv_file:
    reader = csv.reader(csv_file)
    next(reader)  # 跳过列名行（如果CSV文件包含列名行）
    
    # 遍历CSV文件中的每一行
    for row in reader:
        # 提取第四列的内容（假设列索引从0开始），并创建一个字典
        entry = {
            "Origin_reddit": row[1],  # 假设第四列就是我们想要的内容
            "Event": row[1]  # 因示例中没有具体说明Event列的内容，这里使用静态字符串"content"
        }
        # 将字典添加到数据列表中
        data.append(entry)

# 将数据列表保存为JSON格式的文件
with open(output_json_filename, mode='w', encoding='utf-8') as json_file:
    json.dump(data, json_file, indent=4)

print(f"Data has been successfully saved to '{output_json_filename}'.")
```

## 增加ID


```python
import json
import glob
import os

# 定义原始文件夹和目标文件夹
source_folder = '/home/qiang/projects/Digital_mental_health/Dataset/Condidate_sentiment_dataset/2_GAS_SPAS/GAS2SPAS_fewshot_llama3_event'
target_folder = '/home/qiang/projects/Digital_mental_health/Dataset/Condidate_sentiment_dataset/2_GAS_SPAS/GAS2SPAS_fewshot_llama3_event_id'

# 确保目标文件夹存在
os.makedirs(target_folder, exist_ok=True)

# 获取所有json文件的列表
files = glob.glob(os.path.join(source_folder, '*.json'))

# 初始化ID
current_id = 1

for file in files:
    # 读取文件内容
    with open(file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    # 为每个Post添加ID
    for item in data:
        item['ID'] = current_id
        current_id += 1
    
    # 写回修改后的数据到新文件夹中
    # 获取原始文件名
    base_name = os.path.basename(file)
    new_file_name = os.path.join(target_folder, f"modified_{base_name}")
    with open(new_file_name, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
    print(file)

print("所有文件处理完毕。")
```

```python
import os
import json

# 指定两个文件夹的路径
output_folder="/home/qiang/projects/Digital_mental_health/Dataset/Condidate_sentiment_dataset/2_GAS_SPAS/GAS2SPAS_fewshot_llama3_event"
folder1_path = "/home/qiang/projects/Digital_mental_health/Dataset/Condidate_sentiment_dataset/2_GAS_SPAS/GAS_fewshot_llama3"
folder2_path = "/home/qiang/projects/Digital_mental_health/Dataset/Condidate_sentiment_dataset/2_GAS_SPAS/GAS2SPAS_fewshot_llama3"

# 创建一个新的列表来存储新的元素
new_elements = []

# 循环遍历文件夹中的所有文件
for filename in sorted(os.listdir(folder1_path)):
    if filename.endswith(".json"):
        # 读取第一个文件夹中的 JSON 文件
        with open(os.path.join(folder1_path, filename), "r") as f1:
            print(os.path.join(folder1_path, filename))
            data1 = json.load(f1)
        # 读取对应的文件在第二个文件夹中的 JSON 文件
        filename_sas2uas=os.path.splitext(filename)[0]+'_GAS2SPAS.json'
        with open(os.path.join(folder2_path, filename_sas2uas), "r") as f2:
            print(os.path.join(folder2_path, filename_sas2uas))
            data2 = json.load(f2)
        # 创建新的元素
         # 确保两个文件中的元素数量相同
        if len(data1) != len(data2):
            print(f"Error: The number of elements in {filename} does not match between the two folders.")
            continue
        # 遍历第一个文件夹中的每个元素
        for i in range(len(data1)):
            # 获取对应索引位置的元素
            element1 = data1[i]
            element2 = data2[i]
            # 创建新的元素
            new_element = {
                "Event": element1["Event"],
                "Post": element2["Post"],
                "Transferred_Post": element2["Transferred_Post"]
            }
            # 将新的元素添加到列表中
            new_elements.append(new_element)
        # 将新的元素写入到新的 JSON 文件中
        output_filename = os.path.join(output_folder, filename_sas2uas)
        with open(output_filename, "w") as outfile:
            json.dump(new_elements, outfile, indent=4)
```
